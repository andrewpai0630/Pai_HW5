```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
merge_table = pd.merge(city_df, ride_df, on="city", how="outer")
merge_table = merge_table[["city", "date", "fare", "ride_id", "driver_count", "type"]]

# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban = merge_table.groupby(["type"]).get_group("Urban")
urban_avg = urban.groupby(["city"]).mean()["fare"]
urban_cts = urban.groupby(["city"]).count()["fare"]
urban_dc = urban.groupby(["city"]).mean()["driver_count"]
suburban = merge_table.groupby(["type"]).get_group("Suburban")
suburban_avg = suburban.groupby(["city"]).mean()["fare"]
suburban_cts = suburban.groupby(["city"]).count()["fare"]
suburban_dc = suburban.groupby(["city"]).mean()["driver_count"]                                                 
rural = merge_table.groupby(["type"]).get_group("Rural")
rural_avg = rural.groupby(["city"]).mean()["fare"]
rural_cts = rural.groupby(["city"]).count()["fare"]
rural_dc = rural.groupby(["city"]).mean()["driver_count"]

# Build the scatter plots for each city types

plt.scatter(urban_cts, urban_avg, marker="o", facecolor="lightcoral", edgecolors="black", s=10*urban_dc, alpha=.75)
plt.scatter(suburban_cts, suburban_avg, marker="o", facecolor="lightskyblue", edgecolors="black", s=10*suburban_dc, alpha=.75)
plt.scatter(rural_cts, rural_avg, marker="o", facecolor="gold", edgecolors="black", s=10*rural_dc, alpha=.75)

#Incorporate the other graph properties

plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid()

# Create a legend

lgnd = plt.legend(("Urban", "Suburban", "Rural"), markerscale=0.3, title="City Type", loc="best")
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]
plt.savefig("BubblePlot.png")

# Create a text label regarding circle size


```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_ttl = urban.groupby(["type"]).sum()["fare"].rename("Total Fare (Urban)")
suburban_ttl = suburban.groupby(["type"]).sum()["fare"].rename("Total Fare (Suburban)")
rural_ttl = rural.groupby(["type"]).sum()["fare"].rename("Total Fare (Rural)")
# Build Pie Chart
city_type = ["Urban", "Rural", "Suburban"]
total_by_city = [urban_ttl, rural_ttl, suburban_ttl]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(total_by_city, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=285)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
# Save Figure
plt.savefig("TotalFarePieChart.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_cts = urban.groupby(["type"]).count()["city"]
suburban_cts = suburban.groupby(["type"]).count()["city"]
rural_cts = rural.groupby(["type"]).count()["city"]
# Build Pie Chart
city_type = ["Urban", "Rural", "Suburban"]
total_rides_by_city = [urban_cts, rural_cts, suburban_cts]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(total_rides_by_city, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=265)
plt.axis("equal")
plt.title("% of Total Rides by City Type")
# Save Figure
plt.savefig("TotalRidesPieChart.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

urban_dc = urban.groupby(["type"]).sum()["driver_count"]
suburban_dc = suburban.groupby(["type"]).sum()["driver_count"]
rural_dc = rural.groupby(["type"]).sum()["driver_count"]

# Build Pie Charts
city_type = ["Urban", "Rural", "Suburban"]
total_drivers_by_city = [urban_dc, rural_dc, suburban_dc]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(total_drivers_by_city, explode=explode, labels=city_type, colors=colors, autopct="%1.1f%%", shadow=True, startangle=215)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")

# Save Figure
plt.savefig("TotalDriversPieChart.png")
```

```python
# Show Figure
plt.show()
```
